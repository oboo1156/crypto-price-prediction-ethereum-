```python
#libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
#get the dataset from yahoo finance
df = pd.read_csv('downloads/ETH-USD.csv')
df
```

```python
df.dropna(inplace=True)
```

```python
df.isna().any()
```

```python
#create a column that is 'n' days into future
projection = 14 #14days

#prediction column
df['Prediction'] = df[['Close']].shift(-projection)#shift , shifts index by number of periods with opt time frequency
df
```

```python
#independent variable
X = np.array(df[['Close']])

#removing last 14
X = X[:-projection]

X
```

```python
#dependent data
y = df['Prediction'].values
y = y[:-projection]
y
```

```python
#85% train and 15% test splitting
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.15)
```

```python
#modelling
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
#score testing
model.score(X_test,y_test)
```

```python
#setting last shifted rows to x_projection 
x_projection = np.array(df[['Close']][-projection:])
x_projection
```

```python
#prediction
model.predict(x_projection)
```
